```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import numpy as np

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues,
                          figsize=(10, 7),
                          hide_ticks=False):  # Add hide_ticks parameter
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.figure(figsize=figsize)  # Set figsize
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    
    if hide_ticks:
        plt.xticks([])
        plt.yticks([])
    else:
        tick_marks = np.arange(len(classes))
        plt.xticks(tick_marks, classes, rotation=45)
        plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, format(cm[i, j], fmt),
                     ha="center", va="center",
                     color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

# Définir les noms des modèles dans le même ordre que la liste 'models'
model_list = ['GradientBoostingClassifier', 'XGBClassifier', 'RandomForest', 'DecisionTreeClassifier','AdaBoostClassifier','CatBoostClassifier']

# Calculer l'accuracy pour chaque modèle
models = [gb, best_model, best_model_rf, final_model,adaboost,catboost]
accuracy_scores = []

for model in models:
    model.fit(X_train1, y_train1)  # Ajuster le modèle avec les données d'entraînement
    y_pred = model.predict(X_test1)
    accuracy = accuracy_score(y_test1, y_pred)
    accuracy_scores.append(accuracy)

# Trier les modèles par ordre décroissant d'accuracy
sorted_model_names = [name for _, name in sorted(zip(accuracy_scores, model_list), reverse=True)]
sorted_accuracy_scores = sorted(accuracy_scores, reverse=True)

# Afficher les modèles classés du plus performant au moins performant
for name, accuracy in zip(sorted_model_names, sorted_accuracy_scores):
    print(f"{name}: Accuracy = {accuracy}")

# Tracer la courbe d'accuracy du modèle le plus performant
best_model_name = sorted_model_names[0]
best_accuracy = sorted_accuracy_scores[0]

plt.figure()
plt.plot(sorted_model_names, sorted_accuracy_scores, 'ro-', label=f"{best_model_name} (Accuracy = {best_accuracy:.2f})")
plt.xlabel('Modèles')
plt.ylabel('Accuracy')
plt.title('Performance des modèles en termes d\'accuracy')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
